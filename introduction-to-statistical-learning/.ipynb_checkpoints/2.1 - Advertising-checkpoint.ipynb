```python
import pandas as pd
```

```python
df = pd.read_csv('data/Advertising.csv')
df = df.rename(index=str, columns={'Unnamed: 0': 'index'})
df = df.set_index('index')
df.head()
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
def plot_mb(column_name: str):
    X = df[column_name].values.reshape(-1, 1) 
    y = df['sales'].values.reshape(-1, 1)
    reg = LinearRegression().fit(X, y)
    m = reg.coef_[0][0]
    b = reg.intercept_[0]
    plt.scatter(X, y)
    plt.plot(X, [m*x+b for x in X], c='r')
    plt.ylabel('sales')
    plt.xlabel(column_name)
    return m, b
```

```python
plot_mb('TV')
```

```python
plot_mb('radio')
```

```python
plot_mb('newspaper')
```
